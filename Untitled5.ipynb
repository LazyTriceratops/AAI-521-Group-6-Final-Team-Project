```python
# Metadata

metadata_path = os.path.join(path, "WLASL_v0.3.json")
if os.path.exists(metadata_path):
    print(f"Metadata file found: {metadata_path}")
else:
    print("Metadata file not found.")

import json

metadata_file = "/root/.cache/kagglehub/datasets/risangbaskoro/wlasl-processed/versions/5/WLASL_v0.3.json"

# Load and inspect the JSON file
with open(metadata_file, 'r') as f:
    metadata = json.load(f)

# Display the first few entries
print("Number of entries in metadata:", len(metadata))
print("Sample entry:", metadata[0])  # Adjust index if needed

import pandas as pd

# Parse the JSON metadata
metadata_list = []
for entry in metadata:
    gloss = entry['gloss']
    label = entry.get('label', None)  # Add a label if available
    for instance in entry['instances']:
        metadata_list.append({
            "gloss": gloss,
            "label": label,
            "video_id": instance['video_id'],
            "url": instance.get('url', None),
            "bbox": instance.get('bbox', None),
            "fps": instance['fps'],
            "frame_start": instance['frame_start'],
            "frame_end": instance['frame_end'],
            "signer_id": instance['signer_id'],
            "source": instance['source'],
            "split": instance['split']
        })

metadata_df = pd.DataFrame(metadata_list)
print(metadata_df.head())

print("Number of unique glosses:", metadata_df['gloss'].nunique())
print("Number of videos:", metadata_df['video_id'].nunique())

print(metadata_df['split'].value_counts())

gloss_distribution = metadata_df['gloss'].value_counts()
print(gloss_distribution)
```

# **Data PreProcessing**

```python
# Frame Extraction

def extract_and_save_frames(video_id, video_path, output_folder, output_size=(224, 224)):
    # Check if frames are already extracted
    output_file = os.path.join(output_folder, f"{video_id}.npy")
    if os.path.exists(output_file):
        print(f"Frames already extracted for video: {video_id}, skipping...")
        return

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Failed to open video: {video_id}")
        return

    frames = []
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv2.resize(frame, output_size)
        frames.append(frame)
    cap.release()

    # Save frames as .npy file
    if frames:
        os.makedirs(output_folder, exist_ok=True)
        np.save(output_file, np.array(frames))
        print(f"Extracted and saved {len(frames)} frames for video: {video_id}")
    else:
        print(f"No frames extracted from video: {video_id}")


# Define paths
video_folder = "/root/.cache/kagglehub/datasets/risangbaskoro/wlasl-processed/versions/5/videos"
output_folder = "/content/drive/MyDrive/extracted_frames/train"

# Filter metadata for train split
train_videos = valid_metadata_df[valid_metadata_df['split'] == 'train']

# Process each video in the train split
for _, row in train_videos.iterrows():
    video_id = row['video_id']
    video_path = os.path.join(video_folder, f"{video_id}.mp4")

    # Extract and save frames
    extract_and_save_frames(video_id, video_path, output_folder)

# Count extracted frame files
extracted_files = os.listdir(output_folder)
print(f"Number of videos processed: {len(extracted_files)}")
```

```python
# Define paths
video_folder = "/root/.cache/kagglehub/datasets/risangbaskoro/wlasl-processed/versions/5/videos"
output_folder_val = "/content/drive/MyDrive/extracted_frames/val"

# Filter metadata for validation split
val_videos = valid_metadata_df[valid_metadata_df['split'] == 'val']

# Process each video in the validation split
for _, row in val_videos.iterrows():
    video_id = row['video_id']
    video_path = os.path.join(video_folder, f"{video_id}.mp4")

    # Extract and save frames
    extract_and_save_frames(video_id, video_path, output_folder_val)

# Count extracted frame files
extracted_files = os.listdir(output_folder_val)
print(f"Number of videos processed: {len(extracted_files)}")
```

```python
def normalize_frames(input_folder, output_folder):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Normalize each .npy file
    for file_name in os.listdir(input_folder):
        if file_name.endswith(".npy"):
            input_file_path = os.path.join(input_folder, file_name)
            output_file_path = os.path.join(output_folder, file_name)

            # Skip if already normalized
            if os.path.exists(output_file_path):
                print(f"File already normalized: {file_name}, skipping...")
                continue

            # Load frames
            frames = np.load(input_file_path)

            # Normalize pixel values to [0, 1]
            normalized_frames = frames / 255.0

            # Save normalized frames
            np.save(output_file_path, normalized_frames)
            print(f"Normalized and saved: {file_name}")

# Example usage for train, val, and test sets
normalize_frames("/content/drive/MyDrive/extracted_frames/train", "/content/drive/MyDrive/normalized_frames/train")
normalize_frames("/content/drive/MyDrive/extracted_frames/val", "/content/drive/MyDrive/normalized_frames/val")
normalize_frames("/content/drive/MyDrive/extracted_frames/test", "/content/drive/MyDrive/normalized_frames/test")


```
